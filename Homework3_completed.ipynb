# Part 4

```python
female_survival = df[df['Sex'] == 'female']['Survived'].mean()
male_survival = df[df['Sex'] == 'male']['Survived'].mean()
under_16_survival = df[df['Age'] < 16]['Survived'].mean()

print(f"Survival Rate for Females: {female_survival: %}")
print(f"Survival Rate for Males:   {male_survival: %}")
print(f"Survival Rate for Under 16: {under_16_survival: %}")
```

```python
survival_table = df.groupby(['Sex', 'Pclass'])['Survived'].mean()

print(survival_table)
```

#### The data confirms the "Women and Children First" rule, as women and children under 16 had significantly higher survival rates than adult males. However, social class played a massive role. A woman in 1st class had over a 45% chance difference of surviving compared to a woman in 3rd class. Ultimately, the highest chance of survival belonged to 1st-class females (almost 97%), while the lowest was reserved for 3rd class males (approximately 13.5%).
