# Create student names and test score variables and lists

```python
students_list = ["Student 1", "Student 2", "Student 3", "Student 4"] 
tests_list = ["Test 1", "Test 2", "Test 3"]

students_allnames = []
students_allaverage = []

numberofstudents = 4
numberoftestscores_each = 3
```

# Loop for student names (4 times) and their scores (3 times)

```python
for student_name in students_list:
    name = input("Enter student name: ")
    students_allnames.append(name)

    scores_for_this_student = []

    for student_score in tests_list:
        score = float(input(f"Enter score for {name}: "))
        scores_for_this_student.append(score)
```

# Calculating average for one student

```python
student_average = sum(scores_for_this_student) / 3 
students_allaverage.append(student_average)
```

# Calculating class average

```python
class_average = sum(students_allaverage) / 4
```

# Printing heading for table

```python
print("\n==== Class Summary ====\n")
print("Name:      Score\n")
```

# Indexing the lists

```python
index = 0
for name in students_allnames:
    score = students_allaverage[index]
    print(f"{name:<10} {score:.1f}\n")
    index += 1
```

# Printing class average and total number of students

```python
print(f"Class Average: {class_average}\n")
print(f"Total Number of Students: {numberofstudents}")
```
